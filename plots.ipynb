```python
import matplotlib.pyplot as plt

ssf_bpp = [0.1,0.2,0.5]
ssf_psnr = [36.7,38.1,39.9]


pix_bpp = [0.1,0.2,1]
pix_psnr =[29.8,30.3,31.5]
```

```python
plt.plot(ssf_bpp, ssf_psnr, marker='o', label='SSF Compression')

# Plotting rate distortion curve for traditional pixel-based compression
plt.plot(pix_bpp, pix_psnr, marker='o', label='Pixel-Based Compression')

# Adding labels and title
plt.xlabel('Bits Per Pixel (bpp)')
plt.ylabel('PSNR (dB)')
plt.title('Rate Distortion Curves')

# Adding legend
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

```python
plt.savefig('results/rate_distortion_curves.png')
```
